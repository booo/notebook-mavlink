```python
%matplotlib notebook
%pylab inline

from pymavlink import mavutil

connection = mavutil.mavlink_connection("./data.bin")

messages = []

while message := connection.recv_msg():
        messages.append(message)
                                                                                                                                                                         
attitude_messages = list(filter(lambda message: message.get_type() == "ATT", messages))

rolls = list(map(lambda message: message.Roll, attitude_messages))
yaws = list(map(lambda message: message.Yaw, attitude_messages))
pitchs = list(map(lambda message: message.Pitch, attitude_messages))


plot(range(0, len(rolls)), rolls, label='Roll')
plot(range(0, len(yaws)), yaws, label='Yaw')
plot(range(0, len(pitchs)), pitchs, label='Pitch')

xlabel("t")
ylabel("Roll in degree?")
legend(loc='upper left')
```
